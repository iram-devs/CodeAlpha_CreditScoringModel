```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('credit_risk_dataset.csv')
df.head()
```

```python
print("Shape of Dataset:", df.shape)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nSummary Statistics:")
print(df.describe())
```

```python
df['Default'].value_counts().plot(kind='bar')
plt.xlabel('Default (0 = Good Credit, 1 = Bad Credit)')
plt.ylabel('Number of Customers')
plt.title('Credit Risk Distribution')
plt.show()
```

```python
plt.hist(df['Income'], bins=20)
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Distribution')
plt.show()
```

```python
plt.hist(df['Debt'], bins=20)
plt.xlabel('Debt')
plt.ylabel('Frequency')
plt.title('Debt Distribution')
plt.show()
```

```python
sns.boxplot(x='Default', y='Late_Payments', data=df)
plt.title('Late Payments vs Credit Risk')
plt.show()
```

```python
sns.boxplot(x='Default', y='Credit_Utilization', data=df)
plt.title('Credit Utilization vs Credit Risk')
plt.show()
```

```python
corr = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
print("PROJECT CONCLUSIONS")
print("1. Customers with higher debt tend to have a greater chance of default.")
print("2. More late payments increase credit risk.")
print("3. Higher credit utilization is associated with higher default probability.")
print("4. Machine learning models can help classify customers into good and bad credit risks.")
```
