```python
import streamlit as st
import joblib

model = joblib.load('credit_risk_model.pkl')
print(model)
```

```python
new_customer = [[60000, 30, 15000, 0, 25, 4]]

prediction = model.predict(new_customer)

if prediction[0] == 0:
    print("Good Credit Risk")
else:
    print("Bad Credit Risk")
```

```python
st.title("Credit Risk Prediction System")

income = st.number_input("Income")
age = st.number_input("Age")
debt = st.number_input("Debt")
late_payments = st.number_input("Late Payments")
credit_util = st.number_input("Credit Utilization")
employment_years = st.number_input("Employment Years")

if st.button("Predict"):
    data = [[income, age, debt,
             late_payments,
             credit_util,
             employment_years]]

    prediction = model.predict(data)

    if prediction[0] == 0:
        st.success("Good Credit Risk")
    else:
        st.error("Bad Credit Risk")
```
