```python
import numpy as np
import pandas as pd
```

```python
n=1000
df = pd.DataFrame({
    'Income': np.random.randint(20000, 150000, n),
    'Age': np.random.randint(18, 60, n),
    'Debt': np.random.randint(0, 100000, n),
    'Late_Payments': np.random.randint(0, 10, n),
    'Credit_Utilization': np.random.randint(0, 100, n),
    'Employment_Years': np.random.randint(0, 30, n)
})

# Simple rule to create labels
df['Default'] = (
    (df['Debt'] > 50000) &
    (df['Late_Payments'] > 3) &
    (df['Credit_Utilization'] > 70)
).astype(int)

df.to_csv('credit_risk_dataset.csv', index=False)
```
